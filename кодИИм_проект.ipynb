```python
import torch
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import torchvision as tvs
from torch.utils.data import DataLoader
import torchvision.transforms as T
# from torcheval.metrics.functional import multiclass_f1_score
from sklearn.metrics import f1_score
import albumentations as A
from PIL import Image
```

```python
labels = pd.read_csv('/content/Damage_level_Detection-1/train/_annotations.csv')
labels.iloc[50:60]
```

```python
labels['class'] = labels['class'].replace('Minorrotation', 0)
labels['class'] = labels['class'].replace('Moderaterotation', 1)
labels['class'] = labels['class'].replace('Severerotation', 2)
```

```python
labels['class'][labels['class'] == 1].count(), labels['class'][labels['class'] == 2].count(), labels['class'][labels['class'] == 3].count()
```

```python
class Dataset(torch.utils.data.Dataset):
  def __init__(self, data, path='/'):
    self.x_paths = [data.iloc[i, 0] for i in range(len(data))]
    self.labels = [data.iloc[i, 3] for i in range(len(data))]
    self.path = path
  
  def __len__(self):
    return len(self.x_paths)

  def __getitem__(self, idx):
    transform = A.Compose([A.augmentations.transforms.GaussNoise(var_limit = (0, 0.05))])
    img = transform(image = np.array(Image.open(self.path + self.x_paths[idx])) / 255)
    x = torch.Tensor(img['image']).permute(2, 0, 1)
    x = T.functional.normalize(x, [0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    y = self.labels[idx]
    return x, y
```

```python
def torch_imshow(img: torch.Tensor):
  return plt.imshow(img.permute(1, 2, 0))
```

```python
dataset = Dataset(labels, '/content/Damage_level_Detection-1/train/')
```

```python
train_dataset, val_dataset, test_dataset = torch.utils.data.random_split(dataset, [0.7, 0.2, 0.1])
```

```python
train_dataloader = DataLoader(train_dataset, 1, shuffle = True, num_workers = 2)
val_dataloader = DataLoader(val_dataset, 64, shuffle = False, num_workers = 2)
test_dataloader = DataLoader(test_dataset, 64, shuffle = False, num_workers = 2)
```

```python
model = tvs.models.resnet50(weights=tvs.models.ResNet50_Weights.DEFAULT)
```

```python
def pred_one_sample(model, input):
  with torch.no_grad():
    model.eval()
    logits = model(input)
    probs = torch.nn.functional.softmax(logits, dim=-1).numpy()
  return probs
```

```python
DEVICE = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
dataset_sizes = {"train": len(train_dataset), "val": len(val_dataset)}
```

```python
def train_model(model, dataloaders, optimizer, criterion, scheduler, num_epochs=25, device=DEVICE):
  best_model = {'model': 0, 'best_acc': 0}

  for epoch in range(num_epochs):
    print(f'Epoch {epoch} / {num_epochs}')
    print('-' * 10)

    for phase in ['train', 'val']:
      if phase == 'train':
        model.train()
      else:
        model.eval()
      
      running_loss = 0.0
      running_corrects = 0

      for inputs, labels in dataloaders[phase]:
        inputs = inputs.to(device)
        labels = labels.to(device)
        
        
        optimizer.zero_grad()

        with torch.set_grad_enabled(phase == 'train'):
          outputs = model(inputs)
          _, preds = torch.max(outputs, 1)
          print(type(preds))
          print(type(labels))
          # loss = f1_score(outputs.detach().numpy(), labels)
          loss = criterion(preds, labels)

          if phase == 'train':
            loss.backward()
            optimizer.step()
        
        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == labels.data)
      
      if phase == 'train':
        scheduler.step()
      
      epoch_loss = running_loss / dataset_sizes[phase]
      epoch_acc = running_corrects.double() / dataset_sizes[phase]

      print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

      if phase == 'val' and epoch_acc > best_model["best_acc"]:
                best_model['best_acc'] = epoch_acc
                best_model['model'] = model.state_dict()

    print()
    print(f'Best val Acc: {best_model["best_acc"]:4f}')
    
    model.load_state_dict()
    return model
```

```python
optim = torch.optim.AdamW(model.parameters(), lr = 1e-3)
dataloaders = {'train': train_dataloader, 'val': val_dataloader}
scheduler = torch.optim.lr_scheduler.ExponentialLR(optim, 0.9)
```

```python
model = tvs.models.resnet18(weights='IMAGENET1K_V1')
num_ftrs = model.fc.in_features
model.fc = torch.nn.Linear(num_ftrs, 3)
model = model.to(DEVICE)
criterion = torch.nn.CrossEntropyLoss()
```

```python
model = train_model(model, dataloaders, optim, criterion, scheduler)
# model = train_model(model, dataloaders, optim, scheduler)
```
